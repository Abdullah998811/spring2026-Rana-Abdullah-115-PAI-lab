```python
n = int(input("Chessboard size (N): "))
queen_cols = []

print(f"Enter column (0-{n-1}) for each row:")
for row in range(n):
    while True:
        try:
            col = int(input(f"  Row {row} → column: "))
            if 0 <= col < n:
                queen_cols.append(col)
                break
            else:
                print(f"Column must be between 0 and {n-1}")
        except ValueError:
            print("Please enter a number")

is_safe = True

for i in range(n):
    for j in range(i + 1, n):
        if queen_cols[i] == queen_cols[j] or \
           abs(queen_cols[i] - queen_cols[j]) == abs(i - j):
            is_safe = False

print("\n" + "="*40)
print("Valid placement:" if is_safe else "Invalid placement")
print("="*40)

top_border = "+---" * n + "+"
cell = "|   "
queen_cell = "| Q "

print(top_border)
for row in range(n):
    line = ""
    for col in range(n):
        if col == queen_cols[row]:
            line += queen_cell
        else:
            line += cell
    print(line + "|")
    print(top_border)

print()
```
