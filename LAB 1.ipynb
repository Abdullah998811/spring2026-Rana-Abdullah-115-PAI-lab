```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor

train_df = pd.read_csv("C:\\Users\\aimay\\OneDrive\\Desktop\\PAI\\train.csv")
test_df = pd.read_csv("C:\\Users\\aimay\\OneDrive\\Desktop\\PAI\\test.csv")

target = train_df["SalePrice"]

selected_cols = [
    "MSSubClass", "MSZoning", "LotFrontage", "LotArea",
    "Street", "Alley", "LotShape", "LandContour",
    "Utilities", "LotConfig", "LandSlope", "Neighborhood",
    "Condition1", "Condition2", "BldgType", "HouseStyle",
    "OverallQual", "OverallCond", "YearBuilt", "YearRemodAdd",
    "RoofStyle", "RoofMatl", "Exterior1st", "Exterior2nd",
    "MasVnrType", "MasVnrArea", "ExterQual", "ExterCond",
    "Foundation", "BsmtQual"
]

X_train = train_df.loc[:, selected_cols]
X_test = test_df.loc[:, selected_cols]

X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)

X_train, X_test = X_train.align(X_test, axis=1, fill_value=0)

train_means = X_train.mean()
X_train = X_train.fillna(train_means)
X_test = X_test.fillna(train_means)

regressor = DecisionTreeRegressor(random_state=42)
regressor.fit(X_train, target)

predictions = regressor.predict(X_test)

output = pd.DataFrame({
    "Id": test_df["Id"],
    "SalePrice": predictions
})

output.to_csv("submission.csv", index=False)

print("Submission file generated!")
print(f"Number of predictions: {len(output)}")
```
