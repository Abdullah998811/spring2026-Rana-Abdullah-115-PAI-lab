```python
print("Water Jug Puzzle")
print(" 1 → Fill 4L jug")
print(" 2 → Fill 3L jug")
print(" 3 → Empty 4L jug")
print(" 4 → Empty 3L jug")
print(" 5 → Pour 4L → 3L (until source empty)")
print(" 6 → Pour 4L → 3L (until target full)")
print(" 7 → Pour 3L → 4L (until source empty)")
print(" 8 → Pour 3L → 4L (until target full)")
print(" 9 → Quit")

def play_water_jug():
    a = 0          # 4-liter jug
    b = 0          # 3-liter jug
    cap_a = 4
    cap_b = 3
    target = 2

    while True:
        print(f"  4L: {a:2d} | 3L: {b:2d}")
        try:
            choice = int(input("→ "))
        except:
            print("Please enter a number.")
            continue

        if choice == 1:
            a = cap_a
        elif choice == 2:
            b = cap_b
        elif choice == 3:
            a = 0
        elif choice == 4:
            b = 0
        elif choice == 5:   # A → B  (empty A as much as possible)
            can_pour = min(a, cap_b - b)
            a -= can_pour
            b += can_pour
        elif choice == 6:   # A → B  (fill B)
            can_pour = min(a, cap_b - b)
            a -= can_pour
            b += can_pour
        elif choice == 7:   # B → A  (empty B)
            can_pour = min(b, cap_a - a)
            b -= can_pour
            a += can_pour
        elif choice == 8:   # B → A  (fill A)
            can_pour = min(b, cap_a - a)
            b -= can_pour
            a += can_pour
        elif choice == 9:
            print("Goodbye.")
            break
        else:
            print("Wrong option.")
            continue

        if a == target or b == target:
            print(f"  4L: {a:2d} | 3L: {b:2d}")
            print("★ Goal reached! ★")
            break

play_water_jug()
```
